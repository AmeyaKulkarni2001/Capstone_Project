```python
from sklearn.metrics.pairwise import linear_kernel
from sklearn.feature_extraction.text import TfidfVectorizer
import pymongo
import os
import pandas as pd
from dotenv import load_dotenv
load_dotenv()

# Connecting to db
mongo_uri = os.getenv('MONGO_URI')
client = pymongo.MongoClient(mongo_uri)
db = client["TripPlanner"]

# Getting data from db
places_data = db["new_places"].find()

# Creating data frame out of retieved data
new_places_df = pd.DataFrame(list(places_data))
new_places_df
```

```python
# Select the 'features__properties__kinds' column and convert it to a list
tags_list = new_places_df['features__properties__kinds'].tolist()

# Join the list of tags using commas to create a single string
tags_str = ','.join(tags_list)

# Split the string of tags on commas to create a list of individual tags
tags = tags_str.split(',')

# Convert the list of tags to a set to get the unique tags
unique_tags = set(tags)
unique_tags
```
